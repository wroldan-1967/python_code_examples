```python
#################################################
# Script Python para ejecutar en Jupyter Notebook
# Adriana Samacá 
# 15 de Octubre - 2021 
# REESCALACION DE SEñALES 
#################################################

# Importar Clase random para hacer aleatoria la señal generada y Numpy para obtener el promedio en la Re-escalacion final 
import random, numpy as np 
from numpy import array, average

# Clase generadora de señal original con el metodo hacer_aleatoria  el cual [rebaraja n veces con el mismo factor de entrada]  
# los numeros generados originalmente en la lista de tamaño N, esto apra hacer el ejercicio mas realista 
class Generar_lista():
    def __init__(self, ntamaño):
        self.ntamaño = ntamaño
        self.myList = list(range(1, self.ntamaño))
    
    def hacer_aleatoria(self, nveces): 
        for i in range(nveces):
            random.shuffle(self.myList)
        return(self.myList)

# Funcion que obtiene la longitud minima de todas las escalas generadas a partir del factor, sobre la lista original 
def find_min_list(list):
    list_len = [len(i) for i in list]
    return min(list_len)

# Funcion para generar las n Escalas sobre la señal de tamaño (ntamaño) con base en el factor ingresado por el usuario  
def generar_escalas(señal_aleatoria, nfactor):
    result = []
    for i in range(0, len(señal_aleatoria), nfactor):
        slice_item = slice(i, i + nfactor, 1)
        result.append(señal_aleatoria[slice_item])
    print(f'Escalas a partir de la señal, aplicando factor {nfactor} : {result} \n')    
    return result 

# funcion que valdia si la ultima escala no es multiplo del factor y la elimina en caso tal 
def validar_escalas(escala_minima, nfactor, señal):
    if (escala_minima < nfactor):
        print(f'Longitud de escala no valida ({escala_minima}) con valor {señal[-1]} \n')
        larray=len(señal)
        señal.pop(larray-1)
        larray=len(señal)
        print(f'**{larray}**  Escalas Utiles generadas desde la señal con factor {nfactor}  \nEscalas de Señal Antes de Re-Escalación : {señal} \n')
    return señal  

# Funcion que obtiene la señal ya escalada y limpia para proceder a reducirla con re-escalamiento 
def obtener_señal_escalada(señal_aleatoria, nfactor): 
    result = generar_escalas(señal_aleatoria, nfactor)
    escala_minima  = find_min_list(result)
    señal_escalada = validar_escalas( escala_minima, nfactor, result)
    return señal_escalada 

# Funcion que ejecuta la re-escalacion como tal 
def reescalar_señal_final(señal_escalada): 
    row_avg = average(señal_escalada, axis=1)
    row_avg = row_avg.astype(int)
    nueva_señal = row_avg.tolist()
    print(f'Nueva Señal Re-Escalada   : {nueva_señal} \n')
    return nueva_señal  

# Funcion que orquesta el proceso completo 
def procesar_todo(ntamaño, nfactor):
    try:
        lista_origen = Generar_lista(ntamaño)
        señal_aleatoria = lista_origen.hacer_aleatoria(nfactor)
        array_listas = np.array(señal_aleatoria)
        print(f'Señal Original Generada con Tamaño {ntamaño}  : {array_listas} ')
        señal_escalada = obtener_señal_escalada(señal_aleatoria, nfactor)
        señal_final = reescalar_señal_final(señal_escalada)
    except :
        exit(0)

# Funcion que captura los datos de entrada 
def preguntar():
        while (True): 
            Respuesta = input("¿Desea procesar Señal ? (si/no): ")
            if Respuesta.lower() == "si": # lo pasamos a minúsculas para eso el .lower()
                ntamaño = int(input("¿Tamaño de la Señal ? (Entero): "))
                nfactor = int(input("¿Factor de Escala  ? (Entero): "))
                try:
                    int(ntamaño)
                    int(nfactor)
                    it_is = True
                except ValueError:
                    it_is = False
               
                if it_is: 
                    print(f"Tamaño de muestras de Señal : {ntamaño} \n")
                    print(f"factor para Escalas de Señal : {nfactor } \n")
                    procesar_todo(ntamaño, nfactor)
                else:
                    print("Valores de Entrada Incorrectos \n")
                     
            else:
                break

# Ejecuta el llamado 
if __name__ == "__main__":
    try:
        preguntar()
    except:
        exit(0)
```
